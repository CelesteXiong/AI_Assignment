```python
from sklearn import tree
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("BLE_RSSI_dataset\iBeacon_RSSI_Labeled.csv")
```

## 数据预处理

```python
#将date由string对应成number
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['date']=le.fit_transform(data['date'])
data.head()
#将location由string对应成number
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['location']=le.fit_transform(data['location'])
# list(le.classes_)
```

计算每个类别对应多少样本，字典d的keys为类别，对应values为样本数量

```python
d={}
for i in range(data.shape[0]):
    if data['location'][i] in d.keys():
        d[data['location'][i]]+=1
    else:
        d[data['location'][i]]=0
```

## 选择features和labels

```python
from sklearn.model_selection import train_test_split

X=data.drop(['location'],axis=1)

X_mean=X.mean(axis=0) #均值代替缺失值
for row in X:
    for col in X.index:
        if X[row][col]==-200:
            X[row][col]=X_mean[row]
X.head()
```

```python
y=data['location']
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
dmt = tree.DecisionTreeClassifier()
dmt=dmt.fit(X_train,y_train) 

dmt.score(X_test,y_test)#测试测试集得分
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(dmt, X,y, cv=5) #cv参数用于规定将原始数据分成多少份
scores
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
