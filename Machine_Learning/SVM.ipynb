```python
from sklearn import svm
import pandas as pd
import numpy as np
```

## 数据预处理

```python
data=pd.read_csv("BLE_RSSI_dataset\iBeacon_RSSI_Labeled.csv")
data.head()
# print(data.info()) #show the information
```

```python
from sklearn import preprocessing
#将date由string对应成number
le = preprocessing.LabelEncoder()
data['date']=le.fit_transform(data['date'])
#将location由string对应成number

le = preprocessing.LabelEncoder()
data['location']=le.fit_transform(data['location'])

data.head()
```

```python
from sklearn.model_selection import train_test_split

X=data.drop(['location'],axis=1)
X_mean=X.mean(axis=0) #均值代替缺失值
for row in X:
    for col in X.index:
        if X[row][col]==-200:
            X[row][col]=X_mean[row]
X.head()
```

## 用SVM分类

```python
y=data['location']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.01,random_state=0)
clf = svm.SVC(gamma=0.0115, decision_function_shape='ovo')
clt=clf.fit(X_train,y_train)
clf.score(X_test,y_test)#测试测试集得分
```

## 交叉验证

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(clf, X,y, cv=5) #cv参数用于规定将原始数据分成多少份
scores
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
